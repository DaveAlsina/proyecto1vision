```python
import numpy as np
import numba as nb 
import cv2
from scipy import signal
from matplotlib import pyplot as plt
```

```python
def convolve2D(image: np.ndarray, kernel: np.ndarray):
    
    nrows_kernel, ncols_kernel = kernel.shape
    
    # calcula el padding que hay que poner en los lados de
    # izquierda, derecha, arriba y abajo
    padding_lr = np.floor( ncols_kernel/2 ).astype("int16")
    padding_tb = np.floor( nrows_kernel/2 ).astype("int16")
    
    padded_img = cv2.copyMakeBorder(image, borderType=cv2.BORDER_REPLICATE, top= padding_tb,
                                    bottom= padding_tb,
                                    left = padding_lr,
                                    right = padding_lr)
        
    convolved_image = signal.convolve2d(padded_img, kernel)
    
    return convolved_image
```

```python
def make_sobelx_kernel():
    
    return np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
```

```python
def make_sobely_kernel():

    return np.array([[1, 2, 1],[0, 0, 0],[-1, -2, -1]])
```

```python
def sobel(img: np.ndarray):
    
    #convierte la imagen a escala de grises
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #hace el kernel sobel x
    kx = make_sobelx_kernel()

    #hace el kernel sobel y
    ky = make_sobely_kernel()
    
    #aplica el filtro sobel en x e y 
    convolved_img_x = convolve2D(gray, kx)
    convolved_img_y = convolve2D(gray, ky)

    #suma las imagenes filtradas con sobel
    convolved_img = convolved_img_x + convolved_img_y
    
    
    return convolved_img
```

```python
img = cv2.imread("../imgs/img3.jpg") 

%time
plt.imshow(sobel(img), cmap='gray', vmin=0, vmax=255)
```
